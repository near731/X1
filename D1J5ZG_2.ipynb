```
import numpy as np
import scipy.linalg as la

#A=np.array([[0,1],[1,0]])
print("Milyen mátrixot szeretnél invertálni?")

S=int(input("Ird be a sorok számát: "))
O=int(input("Ird be az oszlopok számát: "))

print("Ird be a matrixot: ")
szamok= list(map(int, input().split()))

A=np.array(szamok).reshape(S, O)
print(A)

if (S!=O):
  print("Nem egyezik meg a sorok és az oszlopok száma, igy nem lehet invertálni!")
else:
 if (la.det(A)==0):
   print("Nem lehet a matrixot invertalni")
 else:
   print(la.inv(A)) 

```

3.feladat (matplotlib)

```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1000) / 500 - 1
y = np.arange(1000) / 500 - 1

X, Y = np.meshgrid(x, y)
R = np.sqrt(X**32+ Y**64)
f0 = 1000
k = 50
a = np.sin(np.pi * 2 * (f0 * R + k * R**2 / 2))

fig, axs = plt.subplots(1, 2, figsize=(8, 4), constrained_layout=True)
for ax, interp in zip(axs, ['nearest', 'antialiased']):
    ax.imshow(a, interpolation=interp, cmap='gray')
    ax.set_title(f"interpolation='{interp}'")
plt.show()

```
