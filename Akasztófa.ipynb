```
import numpy as np
import random

print("------AKASZTÓFA JÁTÉK------")

szavak=['python','billentyűzet', 'elefánt','állatorvos','hurokáram','megnyúlás','mechatronika','fogászat','intenzitás']

szo=random.choice(szavak)
print("Találd ki a szót!")
print("Irj be egy betűt, hogy megtudd benne van-e a szóban!")

betuk= ''

eletek_szama=15

while eletek_szama>0:
  hiba=0

  for char in szo:
    if char in betuk:
      print(char)

    else:
      print("_")
      hiba+=1

  if hiba==0:
    print("Nyertél :)")
    print("Ez volt a szo:",szo)
    break

  betu=input("Irj egy betűt: ")

  betuk+=betu

  if betu not in szo:
    eletek_szama-=1
    print("Nincs a szóban ez a betű")
    print("Még",+eletek_szama,"életed maradt.")

    if eletek_szama==0:
      print("Vesztettel!")
```
